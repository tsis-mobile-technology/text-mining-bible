```python
# 텍스트를 불러옵니다.
with open('Korean_news.txt', 'r', encoding='utf8') as f:
    content = f.read()
```

```python
print(content)
```

```python
# 기호 제거하기
import re
filtered_content = re.sub(r'[^\s\w\d]',' ', content)
```

이 코드는 `re` 모듈을 가져오는 것입니다. `re` 모듈은 정규 표현식을 사용하여 문자열을 처리하는 데 사용됩니다. 정규 표현식은 특정 패턴을 찾거나 대체하는 데 사용되며, 문자열의 검색, 추출, 치환 등 다양한 작업을 수행할 수 있습니다.

예를 들어, 다음과 같은 상황에서 `re` 모듈을 사용할 수 있습니다:
- 특정 패턴을 찾아서 문자열에서 추출하고자 할 때
- 문자열에서 특정 패턴을 대체하고자 할 때
- 문자열에서 특정 패턴을 검색하고자 할 때

`re` 모듈을 사용하면 이러한 작업을 간편하게 수행할 수 있습니다.

```python
print(filtered_content)
```

```python
# 키위 형태소 분석기를 사용하기 위해서 Kiwi 클래스 임포트하기
# terminal #> pip install kiwipiepy
from kiwipiepy import Kiwi
kiwi = Kiwi()
```

```python
kiwi_tokens = kiwi.tokenize(filtered_content)
print(kiwi_tokens)
# Tag 정보
# https://github.com/bab2min/kiwipiepy#%ED%92%88%EC%82%AC-%ED%83%9C%EA%B7%B8  
```

각 태그의 품사 정보 <br>
https://github.com/bab2min/kiwipiepy#%ED%92%88%EC%82%AC-%ED%83%9C%EA%B7%B8  

```python
Noun_words = []
for token in kiwi_tokens:
    if 'NN' in token.tag:
        Noun_words.append(token.form)
```

```python
print(Noun_words)
```

명사가 아닌 다른 품사의 단어들을 선택하고자 하는 경우는 아래와 같이 수행

```python
Featured_POS = ['NNG', 'NNP', 'NNB', 'VV', 'VA']
selected_words = []
for token in kiwi_tokens:
    if token.tag in Featured_POS:
        selected_words.append(token.form)
print(selected_words)
```

## Kiwi 에서 제공되는 불용어 사전

```python
from kiwipiepy.utils import Stopwords
stopwords_dict = Stopwords()
print(stopwords_dict.stopwords)
```

### 불용어 제거하기

```python
kiwi.tokenize(filtered_content, stopwords=stopwords_dict)
```

### 불용어 사전에 불용어 추가하기
    선택한 코드는 불용어 사전에 '테스트'라는 단어를 추가하는 코드입니다. 이때, '테스트' 단어는 명사('NNG')로 분류되어 있습니다.

    stopwords_dict.add(('테스트', 'NNG')) 코드는 stopwords_dict라는 집합(set)에 튜플 형태의 불용어를 추가합니다. 여기서 튜플의 첫 번째 요소는 불용어, 두 번째 요소는 그 불용어의 품사를 나타냅니다.

    이렇게 품사 정보를 함께 저장하는 이유는 동일한 단어라도 품사에 따라 그 의미가 달라질 수 있기 때문입니다. 예를 들어, '테스트'라는 단어는 명사로서의 의미와 동사로서의 의미가 다르므로, 품사 정보를 함께 저장함으로써 더 정확한 텍스트 처리를 가능하게 합니다.

```python
stopwords_dict.add(('테스트', 'NNG'))
```

```python
customized_stopwords = ['연합뉴스', '네이버', '기자']
for word in customized_stopwords:
    stopwords_dict.add((word, 'NNG'))

# 불용어를 추가한 다음에는 tokenize() 함수를 다시 호출해야 합니다.
```
